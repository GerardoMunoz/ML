Comenzaremos prediciendo el avance de la diabetes basado en el IMC. 

Matemáticamente hablando, debemos encontrar una función que recibe el IMC y devuelve  el avance de la diabetes.
* IMC es un número real,
* El avance también es un número real.

Este tipo de problemas se conoce como **regresión**.

Para solucionar este problema se utilizará el algoritmo de los vecinos cercanos.
> https://scikit-learn.org/stable/modules/neighbors.html#regression



```python
import numpy as np
import sklearn
```

```python
from sklearn import datasets
```

```python
X_pre, y_pre = datasets.load_diabetes(return_X_y=True)
print(X_pre.shape,y_pre.shape)
print(X_pre[0])
```

```python
print(y_pre[0])
```

```python
X_tmp = X_pre[:, 2] #Selecciona solamente la columna 2 que tiene el IMC
X_tmp.shape
```

```python
X = X_tmp[:, np.newaxis]
X.shape
```

```python
min(X),max(X)
```

```python
from sklearn import model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y_pre, test_size=0.33)
X_train.shape, X_test.shape, y_train.shape, y_test.shape 
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_train, y_train,  color='black', s=1)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor(n_neighbors=1)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred.shape
```

```python
plt.scatter(X_train, y_train,  color='black', s=1)
plt.scatter(X_test, y_pred,  color='yellow', s=1)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

```python
def error(u,v):
  error=0
  for u_i,v_i in zip(u,v):
    error += (u_i - v_i)**2
  return error/len(u)
 
error(y_test, y_pred)
```

```python
from sklearn.neighbors import KNeighborsRegressor
model_10 = KNeighborsRegressor(n_neighbors=10)
model_10.fit(X_train, y_train)
y_pred_10 = model_10.predict(X_test)

plt.scatter(X_train, y_train,  color='black', s=1)
plt.scatter(X_test, y_pred_10,  color='yellow', s=1)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()

mean_squared_error(y_test, y_pred_10)
```

```python
from sklearn.neighbors import KNeighborsRegressor
model_40 = KNeighborsRegressor(n_neighbors=40)
model_40.fit(X_train, y_train)
y_pred_40 = model_40.predict(X_test)

plt.scatter(X_train, y_train,  color='black', s=1)
plt.scatter(X_test, y_pred_40,  color='yellow', s=1)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()

mean_squared_error(y_test, y_pred_40)
```
